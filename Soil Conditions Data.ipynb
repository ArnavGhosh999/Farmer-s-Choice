```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/archive.zip')
```

#N - ratio of Nitrogen content in soil
#P - ratio of Phosphorous content in soil
#K - ratio of Potassium content in soil
#temperature - temperature in degree Celsius
#humidity - relative humidity in %
#ph - ph value of the soil
#rainfall - rainfall in mm

###The first dataset tells about what crop grows in conditions such as Nitrogen, Phosphorous, Potassium levels, having what pH and the optimum rainfall needed with the optimum temperature and humidity.

```python
df
```

```python
df.groupby('LABEL')['TEMPERATURE'].mean().plot(kind='bar', color='g')
plt.xlabel('Crop Type',fontsize='15')
plt.ylabel('Temperature',fontsize='15')
```

```python
df.columns
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='PH', y='RAINFALL', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.columns
```

```python
sns.scatterplot(data=df, x='TEMPERATURE', y='RAINFALL', hue='LABEL')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(data=df, x='LABEL', y='TEMPERATURE')
plt.xlabel('Crop Type', fontsize=15)
plt.ylabel('Temperature', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, x='LABEL')
plt.xlabel('Crop Type', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.histplot(data=df, x='RAINFALL', bins=20, kde=True)
plt.xlabel('Rainfall', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.show()
```
