```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/archive.zip')
```

```python
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Production', y='Annual_Rainfall', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.plot.scatter(x='Annual_Rainfall', y='Yield')
```

```python
sns.lineplot(data=df, x='Crop_Year', y='Yield', hue='Crop')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from matplotlib import pyplot as plt
df['Annual_Rainfall'].plot(kind='hist', bins=20, title='Annual_Rainfall')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Season'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Crop_Year', y='Season', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
df.groupby('State')['Yield'].mean().plot(kind='bar',figsize=(10,8))
plt.xlabel('STATE')
plt.ylabel('YIELD')
```

```python
from matplotlib import pyplot as plt
df['Crop_Year'].plot(kind='line', figsize=(8, 4), title='Crop_Year')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Season", y="Yield", hue="State")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df.groupby(['Crop', 'Crop_Year'])['Production'].sum().unstack().plot(kind='line')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
